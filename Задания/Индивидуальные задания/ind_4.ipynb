>Найти какое-либо изображение в сети Интернет. Создать ноутбук, в котором будет отображено выбранное изображение средствами библиотеки matplotlib по URL из сети Интернет.

Отобразим изображение пейзажа с сайта [oir.mobi](https://oir.mobi/680497-priroda-umirotvorenie-i-spokojstvie.html "Природа, Умиротворение и Спокойствие")

```python
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots()
plt.figure(figsize=(20, 20))

response = requests.get('https://oir.mobi/uploads/posts/2021-06/thumbs/1623788314_6-oir_mobi-p-priroda-umirotvorenie-i-spokoistvie-prirod-7.jpg')
img = Image.open(BytesIO(response.content))

ax.set_title("Умиротворение в природе",
            fontsize=18)

ax.imshow(img)
```
